```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
leitura=pd.read_csv('/kaggle/input/bank-customers/Churn Modeling.csv')
leitura
```

```python
leitura.info()
```

```python
novo_geography=pd.get_dummies(leitura['Geography'],prefix='geography')
novo_gender=pd.get_dummies(leitura['Gender'],prefix='gender')
leitura=leitura.drop(columns=['Geography','Gender','Surname'], axis=1)
leitura=pd.concat([leitura,novo_geography,novo_gender],axis=1)
```

```python
leitura.columns
```

```python
leitura
```

```python
for coluna in leitura: 
    grafico=px.histogram(leitura,x=coluna,color='Exited')
    grafico.show()
```

**Conclusão**

* A saída é maior para o público feminino com 1139 de 2037 saídas;
* São clientes que possuem apenas um produto com a instituição bancária;
* Com faixa etária entre os 35 à 52 anos;
* Nesse caso poderiamos fazer uma campanha e oferecer algum outro produto para esses clientes, tendo em vista, que as saídas de clientes com pelo menos dois produtos é bem menor.
